```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a new column
labels = ['Young Adult', 'Adult', 'Middile-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# craete column purchage frequency days
frequncy_mapping = { 
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Annually' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Monthly' : 365,
    'Every 3 Months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequncy_mapping)
```
